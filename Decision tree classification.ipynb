```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#load dataset
```

```python
dataset = load_iris()
```

```python
df = sns.load_dataset('iris')
```

```python
dataset.target
```

```python
df.head()
```

```python
# Divide data into dependent and independent 
```

```python
X = df.iloc[:,:-1]
X
```

```python
y = dataset.target
```

```python
# train test split implementation
# Decision tree doesnot need standarization or normailzation of value as it will make model more complex
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,random_state = 42)
```

```python
# Import Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#Post pruning - first will make decision tree then we will cut from the branches
# good for small dataset

treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train,y_train)
```

```python
# tree display
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
# decision tree classification with post pruning
treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train,y_train)
```

```python
# tree display
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
# Prediction
y_pred = treeclassifier.predict(X_test)
y_pred
```

```python
# test classification report and accuracy score
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred)*100 ,'percent') 
print(classification_report(y_test,y_pred))
```

```python
# New dataset
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
dataset
```

```python
df = pd.DataFrame(data = dataset.data,columns = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
        'mean smoothness', 'mean compactness', 'mean concavity',
        'mean concave points', 'mean symmetry', 'mean fractal dimension',
        'radius error', 'texture error', 'perimeter error', 'area error',
        'smoothness error', 'compactness error', 'concavity error',
        'concave points error', 'symmetry error',
        'fractal dimension error', 'worst radius', 'worst texture',
        'worst perimeter', 'worst area', 'worst smoothness',
        'worst compactness', 'worst concavity', 'worst concave points',
        'worst symmetry', 'worst fractal dimension'])
```

```python
df.shape
```

```python
X_= df
y_= dataset.target
```

```python
 #Decision tree doesnot need standarization or normailzation of value as it will make model more complex
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_,y_,test_size=0.40,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier_tree = DecisionTreeClassifier()
```

```python
classifier_tree.fit(X_train,y_train)
```

```python
# tree display
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(classifier_tree,filled=True)
```

```python
classifier_tree = DecisionTreeClassifier(max_depth=2)
classifier_tree.fit(X_train,y_train)
```

```python
# tree display
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(classifier_tree,filled=True)
```

```python
y_pred = classifier_tree.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_test,y_pred)*100,'percent') 
print(classification_report(y_test,y_pred))
```
