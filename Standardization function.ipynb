```python
import numpy as np
```

```python
def standardize_function(data):
    # for standardization mean = 0, std = 1

    mean = np.mean(data,axis=0)
    std = np.std(data,axis=0)

    #formula for standardization (X - mean)/std , X - data point
    standardized_data = (data-mean)/std
    return standardized_data
```

```python
def euclidean_distance(X1,X2,y1,y2):
    d = np.sqrt((X2-X1)**2 + (y2-y1)**2)
    return d
    
```

```python
print(euclidean_distance(1,4,2,6))
```

```python
def manhatten_distance(X1,X2,y1,y2):
    d = abs(X1-X2) + abs(y1-y2)
    return d
    
```

```python
print(manhatten_distance(1,4,2,0))
```
